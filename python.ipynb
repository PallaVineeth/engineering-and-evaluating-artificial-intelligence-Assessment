# Helper Functions

```python
def fit(model_type, X, y):
        models = []
        for i in range(y.shape[1]):
            model = model_type()
            if i > 0:
                X = _add_previous_predictions(X, y[:, i-1])
            model.fit(X, y[:, i])
            models.append(model)
        return models
```

```python
def _add_previous_predictions(X, y_prev):
        return np.hstack((X, y_prev.reshape(-1, 1)))
```

```python
def predict(X,models):
        predictions = []
        for model in models:
            pred = model.predict(X)
            predictions.append(pred)
            X = _add_previous_predictions(X, pred)
        return np.column_stack(predictions)
```

```python
def score(X, y, models):
        predictions = predict(X, models)
        accuracy = _calculate_accuracy(predictions, y)
        return accuracy
```

```python
def _calculate_accuracy(predictions, true_labels):
        correct = np.all(predictions == true_labels, axis=1)
        accuracy = np.mean(correct)
        return accuracy
```
